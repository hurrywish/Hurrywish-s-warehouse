```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from scipy import linalg
%matplotlib inline
```

```python
def line(e,s,lab_range):
    e=e/np.linalg.norm(e)
    lambda_1=np.linspace(-lab_range,lab_range,1000)
    L=np.array([s+j*e for j in lambda_1])
    fig,ax=plt.subplots(1,1,figsize=(12,10))
    ax.plot(L[:,0],L[:,1],ls='--',marker='*',color='r',lw=0.5)
    return 
```

```python
line(np.array([12,6]),np.array([2,1.2]),15)
```

```python
from sklearn.datasets import make_regression 
```

```python
data_set=make_regression(1000,1,1,1,0.3,noise=1.2,coef=True,random_state=1)
```

```python
X=np.ravel(data_set[0])
Y=np.ravel(data_set[1])
X.shape,Y.shape
```

```python
m=((X-X.mean())@(Y-Y.mean()))/(np.linalg.norm(X-X.mean())**2)
lambda X,Y: Y.mean()-m*X.mean(),m
```

```python
def line(X,Y):
    alpha_1=((X-X.mean())@(Y-Y.mean()))/(np.linalg.norm(X-X.mean())**2)
    alpha_0=Y.mean()-alpha_1*X.mean()
    y=alpha_0+X*alpha_1
    plt.figure(figsize=(12,10))
    plt.plot(X,Y,label='original')
    plt.plot(X,y,label='correction')
    plt.legend()
    return alpha_0,alpha_1
```

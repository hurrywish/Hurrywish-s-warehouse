```python
from matplotlib.colors import ListedColormap
from sklearn.datasets import make_circles,make_blobs,make_moons,make_classification
```

```python
n_samples=1000

samples={'circles':make_circles(n_samples=n_samples,noise=0.1,factor=0.5,random_state=0),
        'blobs':make_blobs(n_samples=n_samples,centers=2,random_state=0),
        'moons':make_moons(n_samples=n_samples,noise=0.1, random_state=0),
        'classification':make_classification(n_samples=n_samples,n_features=2,n_informative=2,n_redundant=0,random_state=0)
        }

kernel=['none','linear','poly','rbf','sigmoid']


plt.figure(figsize=(14,10))
plt.subplots_adjust(hspace=.5,wspace=0.5)

index=0
for j in kernel:
    for i in samples:
        data=samples[i][0]
        label=samples[i][1]
        x=data[:,0]
        y=data[:,1]
        if j == 'none':
            plt.subplot(5,4,index+1)
            plt.scatter(x,y,c=label,s=1)
            plt.title(i)
        else:
            plt.subplot(5,4,index+1)
            plt.scatter(x,y,c=label,s=1,zorder=10)
            ax=plt.gca()
            
            clf=SVC(kernel=j,random_state=0)
            clf.fit(data,label)
            
            x_mesh=np.linspace(np.min(x),np.max(x),100)
            y_mesh=np.linspace(np.min(y),np.max(y),100)
            x, y=np.meshgrid(x_mesh,y_mesh)
            xy=np.vstack([np.ravel(x),np.ravel(y)]).T
            #plt.subplot(5,4,index+1)
            
            z=clf.decision_function(xy).reshape(100,100)
            ax.contour(x,y,z,levels=[-1,0,1],linestyles=['--','-','--'],colors='k')
            
            ax.pcolormesh(x,y,z>0,cmap=plt.cm.Paired,shading='auto',)
            score=clf.score(data,label)
            plt.title('kernel: %s, score: %0.2f' % ( j,score))
            
        index+=1
            
            
        

    
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
import numpy as np
import datetime
from sklearn.decomposition import PCA
```

```python
data=load_breast_cancer()
sample=data.data
label=data.target
```

```python
np.unique(label)
```

```python
plt.scatter(sample[:,0],sample[:,1],c=label)
```

```python
x_decomp=PCA(n_components=2,random_state=0).fit_transform(sample,label)
x_decomp
```

```python
plt.scatter(x_decomp[:,0],x_decomp[:,1],c=label)
```

```python
x_train,x_test,y_train,y_test=train_test_split(sample,label,train_size=0.8,random_state=0)
```

```python
kernels=['linear','poly','rbf','sigmoid']

for kernel in kernels:
    time0=datetime.datetime.now()
    clf=SVC(kernel=kernel,random_state=0,cache_size=10240,degree=3)
    clf.fit(x_train,y_train)
    test_score=clf.score(x_test,y_test)
    train_score=clf.score(x_train,y_train)
    time1=datetime.datetime.now()
    time=time1-time0
    print('kernel: %s, train_score: %0.2f, test_score: %0.2f, time: %s' %(kernel,train_score, test_score,time))
```

```python
class_1=500
class_2=50

centers=[[0,0],[2,2]]
clusters_std=[1.5,0.5]
x,y=make_blobs(n_samples=[class_1,class_2],centers=centers,cluster_std=clusters_std,random_state=0,shuffle=False)

clf_svc=SVC(kernel='linear',probability=True,C=1,random_state=0)
clf_svc.fit(x,y)
```

```python
clf_svc.predict_proba(x)
```

```python
clf_svc.decision_function(x).shape
```

```python
CM=confusion_matrix(y,clf_svc.predict(x),labels=[1,0])
CM
```

```python
# FPR

CM[1,0]/CM[1,:].sum()
```

```python
# Recall

CM[0,0]/CM[0,:].sum()
```

```python
recall=list()
fpr=list()

probrange=np.linspace(clf_svc.predict_proba(x)[:,1].min(),clf_svc.predict_proba(x)[:,1].max(),50,endpoint=False)

for i in probrange:
    y_predict=list()
    for j in range(len(x)):
        if clf_svc.predict_proba(x)[j,1]>i:
            y_predict.append(1)
        else:
            y_predict.append(0)
            
    cm=confusion_matrix(y,y_predict,labels=[1,0])
    recall.append(cm[0,0]/cm[0,:].sum())
    fpr.append(cm[1,0]/cm[1,:].sum())

recall.sort()
fpr.sort()
```

```python
plt.plot(fpr,recall,c='r')
plt.plot(probrange+0.05,probrange+0.05,c='k',linestyle='--')
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
```

```python
fpr,recall,threshold=roc_curve(y,clf_svc.decision_function(x))
area=roc_auc_score(y,clf_svc.decision_function(x))
area
```

```python
plt.figure(figsize=(12,10))
plt.plot(fpr,recall,color='r',label='ROC curve (area=%0.2f)'%(area))
plt.plot([0,1],[0,1],c='k',ls='--')
plt.xlabel('FPR')
plt.ylabel('Recall')
plt.title('Receiver operating characteristic example')
plt.legend(loc='lower right')

maxindex=(recall-fpr)
maxindex=np.argsort(maxindex)[-1]
plt.scatter(fpr[maxindex],recall[maxindex],s=50,color='b')
plt.text(fpr[maxindex]+0.01,recall[maxindex]+0.01,round(threshold[maxindex],3))
```

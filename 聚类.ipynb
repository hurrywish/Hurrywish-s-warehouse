```python
from sklearn.datasets.samples_generator import make_blobs
```

```python
x,y_true=make_blobs(n_samples=300,centers=4,cluster_std=[0.3,0.5,0.7,0.9],random_state=1)
x,y_true
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=4)
kmeans.fit(x)
y_kmeans=kmeans.predict(x)
y_kmeans
centroids=kmeans.cluster_centers_
```

```python
plt.scatter(x[:,0],x[:,1],c=y_kmeans,cmap='Dark2',s=10,marker='x')
plt.scatter(centroids[:,0],centroids[:,1])
```

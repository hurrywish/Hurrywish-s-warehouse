```python
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
a=np.random.randn(4,5)
b=np.random.randn(5,7)
a,b
```

```python
a@b #内积
```

```python
c=np.random.randn(3)
d=np.random.randn(3)
c,d
```

```python
np.cross(c,d) #叉积
```

```python
origin=np.array([0,0,0])
x=np.array([1,2,0])
y=np.array([4,5,6])
z=np.cross(x,y)
line_x=np.column_stack([origin,x])
line_y=np.column_stack([origin,y])
line_z=np.column_stack([origin,z])

fig1=plt.figure(figsize=(10,10))
ax=Axes3D(fig1)
ax.plot(line_x[0],line_x[1],line_x[2],label='line_x')
ax.plot(line_y[0],line_y[1],line_y[2],label='line_y')
ax.plot(line_z[0],line_z[1],line_z[2],label='line_z')
plt.legend()
```

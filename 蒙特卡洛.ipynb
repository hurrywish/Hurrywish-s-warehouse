```python
n=10000
xmin=0
xmax=1
ymin=0
ymax=1

x=np.random.uniform(xmin,xmax,n)
y=np.random.uniform(ymin,ymax,n)
x_y_dots=pd.DataFrame({'x':x,'y':y})
plt.scatter(x_y_dots['x'],x_y_dots['y'],s=0.3,c='b')
plt.xlim([0,1])
plt.ylim([0,1])

def f(x):
    return x**2

xi=np.linspace(0,1,100)
yi=f(xi)

plt.scatter(x_y_dots['x'],x_y_dots['y'],s=0.3)
plt.xlim([0,1])
plt.ylim([0,1])
plt.plot(xi,yi,c='r')

dots=np.where(x_y_dots['y']<f(x_y_dots['x']),1,0).sum()
dots
dots/n
```

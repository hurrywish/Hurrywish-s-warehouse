```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline
```

```python
df=pd.DataFrame({'value1':np.random.rand(10)*20,'value2':np.random.rand(10)*100})
df.head(10)
```

```python
def guiyi(data,*cols):
    for col in cols:
        col_max=data[col].max()
        col_min=data[col].min()
        value=(data[col]-col_min)/(col_max-col_min) 
    return value

df['value1_01']=guiyi(df,'value1')
df['value2_01']=guiyi(df,'value2')
df
```

```python
def z_score(data,*cols):
    for col in cols:
        col_mean=data[col].mean()
        col_std=data[col].std()
        value=(data[col]-col_mean)/col_std
    return value
```

```python
df['value1_z']=z_score(df,'value1')
df['value2_z']=z_score(df,'value2')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('seaborn-whitegrid')
plt.rcParams['font.sans-serif']=['SimHei'] #选择字体
plt.rcParams.update({'font.size': 12}) #字体大小
```

```python
data=pd.DataFrame(np.random.rand(30,2)*1000,columns=list('AB'),index=pd.date_range('2017-6-1',periods=30))
```

```python
plt.figure(figsize=(10,6))
plt.plot(data['A'],lw=1,ls='--')
plt.plot(data['B'],lw=1,ls='-.')
```

```python
data.plot.bar(stacked=False,figsize=(20,8))
```

```python
x=range(len(data))
y1=data['A']
y2=-data['B']
fig=plt.subplots(figsize=(10,8))
plt.subplots_adjust(hspace=0)
ax1=plt.subplot(311)
ax1.bar(x,y1,color='r')
ax2=plt.subplot(312)
ax2.bar(x,y2,color='b')
fig=plt.subplots(figsize=(10,8))
plt.plot(x,y1+y2,marker='.',ls='-.')
plt.axhline(0,color='blue',linewidth=1)
plt.set_(data.index[::3])
```

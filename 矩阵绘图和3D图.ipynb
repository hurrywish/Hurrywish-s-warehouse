```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from matplotlib import gridspec
%matplotlib inline
```

```python
plt.figure(figsize=(12,12))
gs=gridspec.GridSpec(3,3)
pic1=plt.subplot(gs[0,:2])
pic2=plt.subplot(gs[0,-1])
pic3=plt.subplot(gs[1:,:-1])
pic4=plt.subplot(gs[1,-1])
pic5=plt.subplot(gs[2,-1])
plt.subplots_adjust(wspace=0.2,hspace=0.2)
plt.suptitle('Python_Practice')
#subpicnr_invisible(plt.gcf())

n1=1000
X=np.linspace(-2*np.pi,2*np.pi,n1)
Y=np.sin(2*X)*X+np.pi
Y1=np.sin(2*X)*X-np.pi
pic1.plot(X,Y,ls='--',c='#FFB6C1',alpha=1)
pic1.plot(X,Y1,ls='-.',c='#1E90FF',alpha=1)

n2=1000
X2=np.random.normal(0,1,n2)
Y2=np.random.normal(0,1,n2)
pic2.scatter(X2,Y2,c=np.linspace(0,1,1000),cmap='coolwarm')

n3=10
X3=np.arange(n3)
Y3=(1-X3/np.float(n3))*np.random.uniform(0,0.5,n3)
Y3_1=-Y3
pic3.bar(X3,Y3,width=0.6,align='center',facecolor='#FFB6C1')
pic3.bar(X3,Y3_1,width=0.6,align='center',facecolor='#1E90FF')

for x,y,y1 in zip(X3,Y3,Y3_1):
    pic3.text(x,y+0.01,'%.2f'%y,ha='center')
    pic3.text(x,y1-0.02,'%.2f'%y1,ha='center')

n4=100
f=lambda x,y: (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)
x4=np.linspace(-np.pi,np.pi,n4)
y4=np.linspace(-np.pi,np.pi,n4)
X4,Y4=np.meshgrid(x4,y4)
pic4.contourf(X4,Y4,f(X4,Y4),10,cmap='coolwarm')
```
